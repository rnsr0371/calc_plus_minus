```python
#必要なライブラリのインポート
from selenium import webdriver
```

```python
#半自動の要素１つ目。データを取得するページのURLはコピペする必要がある。
#9/30の琉球vs.A東京の試合のデータにアクセス
url="https://www.bleague.jp/game_detail/?ScheduleKey=7440"
driver=webdriver.Chrome("クロームドライバーのパスに置き換える")
driver.get(url)
```

```python
#プレイバイプレイからプレーの内容を取得
playbyplay=[]
for play in driver.find_elements_by_class_name("detail"):
    playbyplay.append(play.get_attribute("innerHTML"))
```

```python
#プレイバイプレイの順番を逆にする（試合開始に近い順番に並べ替え）
playbyplay.reverse()
playbyplay
```

```python
#プレイごとのラインナップを記録する配列（line_up)を作る
on_court=[]
line_up=[]
for i in range(len(playbyplay)):
    #プレイヤーの交代は「#背番号　選手の名前　プレイヤーイン（/アウト）」とフォーマットが決まっているため、このやり方でOK。
    if "プレイヤーイン" in playbyplay[i]:
        on_court.append(playbyplay[i].split(" ")[1])
    if "プレイヤーアウト" in playbyplay[i]:
        on_court.remove(playbyplay[i].split(" ")[1])
        
    line_up.append(list(on_court))#listがないと最後のメンバーで固定されてしまう
line_up
```

```python
#プレーの内容（playbyplay）とラインナップ（line_up）を結合する。第一要素がプレー内容、第二要素がラインナップ（配列）
data=[]
for p,l in zip(playbyplay,line_up):
    data.append([p,l])
data
```

```python
#半自動の要素２つ目はこれ。Bリーグ公式サイトのPlayByPlayのdetailには所属チームの記載がないので、チームのロスターを記憶させる必要がある。
#この方法では名字が同じ選手がいると、プラスマイナスの値が正確に算出できないので注意。
ryukyu=["小寺","並里","フリッピン","ダーラム","エバンス","岸本","田代","今村","満原","小野寺","クーリー","牧"]
tokyo=["小島","吉井","安藤","バランスキー","サイズ","菊地","ロシター","田中","笹倉","カーク","小酒部"]
```

```python
#dataに保存されているプレー内容がどのチームのものだったのかを記録する配列(team)を作る
team=[]
for i in range(len(data)):
    for r in ryukyu:
        if r in data[i][0]:
            team.append("琉球")
    for t in tokyo:
        if t in data[i][0]:
            team.append("東京")
    if "タイムアウト" in data[i][0]:
        team.append("その他")
        
    if "チーム" in data[i][0]:
        team.append("その他")
        
team
```

```python
#data（プレー内容、オンコート選手）とteam(どのチームのプレイか？)を結合してdata2とする。
data2=[]
for i in range(len(data)):
    data2.append([data[i][0],team[i],data[i][1]])
    
data2#第一要素がプレー内容、第二要素がチーム名、第３要素がオンコート選手
```

```python
#プラスマイナスを記録する辞書を作成する。キーは選手名、値はプラスマイナス。
plus_minus={}
for l in line_up:
    for p in l:
        plus_minus[p]=0
plus_minus
```

```python
#プラスマイナスの算出。プレー内容とチームとオンコート選手の情報を使って、plus_minusという辞書の値を更新していく。
for i in range(len(data2)):
    if "○" in data2[i][0]:
        if "2P" in data2[i][0]:
            if "琉球"==data2[i][1]:
                for p in data2[i][2]:
                    if p in ryukyu:
                        plus_minus[p]+=2
                    else:
                        plus_minus[p]-=2
            if "東京"==data2[i][1]:
                for p in data2[i][2]:
                    if p in ryukyu:
                        plus_minus[p]-=2
                    else:
                        plus_minus[p]+=2
                        
        if "3P" in data2[i][0]:
            if "琉球"==data2[i][1]:
                for p in data2[i][2]:
                    if p in ryukyu:
                        plus_minus[p]+=3
                    else:
                        plus_minus[p]-=3
            if "東京"==data2[i][1]:
                for p in data2[i][2]:
                    if p in ryukyu:
                        plus_minus[p]-=3
                    else:
                        plus_minus[p]+=3
                        
        if "フリースロー" in data2[i][0]:
            if "琉球"==data2[i][1]:
                for p in data2[i][2]:
                    if p in ryukyu:
                        plus_minus[p]+=1
                    else:
                        plus_minus[p]-=1
            if "東京"==data2[i][1]:
                for p in data2[i][2]:
                    if p in ryukyu:
                        plus_minus[p]-=1
                    else:
                        plus_minus[p]+=1
```
